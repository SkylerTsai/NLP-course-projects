## Model

### QA Model

```python
from transformers import BertForMultipleChoice, BertTokenizer
import torch.nn as nn

class bert_classifier(nn.Module):
    def __init__(self, num_class=NUM_CLASS, pretrain_model="bert-base-uncased"):
        super(bert_classifier, self).__init__()
        self.pretrain_model = pretrain_model

        self.num_class = num_class

        self.bert = BertForMultipleChoice.from_pretrained(pretrain_model)
        self.tokenizer = BertTokenizer.from_pretrained(pretrain_model)

        print("BERT config", self.bert.config)

    def forward(self, input_ids, attention_mask, token_type_ids, choice_feature=None, similarity_feature=None, length_feature=None):
        #print(input_ids.size(), attention_mask.size(), token_type_ids.size())
        #print(choice_feature.size(), similarity_feature.size(), length_feature.size())

        output = self.bert(
            input_ids = input_ids,
            attention_mask = attention_mask,
            token_type_ids = token_type_ids,
            output_hidden_states = True,
            return_dict = True)

        #print((output.hidden_states[-1]).size(0))
        #lhs = output.hidden_states[-1] # text1 <CLS>: lhs[0][0]

        return output.logits

    def get_num_class(self):
        return self.num_class

    def get_pretrain_model_name(self):
        return self.pretrain_model.split('/')[-1]

    def get_tokenizer(self):
        return self.tokenizer
```

```python
model = bert_classifier().to(device)

tokenizer = model.get_tokenizer()
loss_fn = torch.nn.CrossEntropyLoss(ignore_index = -1).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr = LEARNING_RATE)
```

### MLP Model

```python
import torch.nn as nn

class mlp_classifier(nn.Module):
    def __init__(self, num_class=NUM_CLASS, use_lstm=False, mode="CESL"):
        super(mlp_classifier, self).__init__()
        self.num_class = num_class

        self.feature_size = 384 + 2 * num_class
        if "C" not in mode: self.feature_size -= num_class
        if "E" not in mode: self.feature_size -= 384
        if "S" not in mode: self.feature_size -= num_class - 1
        if "L" not in mode: self.feature_size -= 1
        print("feature size =", self.feature_size)

        self.use_lstm = use_lstm

        self.mode = mode

        self.linear = nn.Sequential(
            nn.Linear(self.feature_size, 1024),
            nn.ReLU(),
            nn.Dropout(p = 0.5),
            nn.Linear(1024, 1024),
            nn.ReLU(),
            nn.Dropout(p = 0.5),
            nn.Linear(1024, 1),
        )

        self.lstm = nn.LSTM(input_size = BATCH_SIZE,
	        hidden_size = BATCH_SIZE * num_class,
			num_layers  = 16,
			batch_first = True,
			dropout = 0.5,
			bidirectional = False
		)

    def forward(self, choice_feature, embedding_feature, similarity_feature, length_feature):
        #print(choice_feature.size(), embedding_feature.size(), similarity_feature.size(), length_feature.size())

        feature_list = []
        #print(self.mode)
        if "C" in self.mode:
            feature_list.append(choice_feature)
        if "E" in self.mode:
            feature_list.append(embedding_feature)
        if "S" in self.mode:
            feature_list.append(similarity_feature)
        if "L" in self.mode:
            feature_list.append(length_feature)

        features = torch.concat(feature_list, dim = -1).type_as(embedding_feature)
        #print(features.size(), self.feature_size)
        #print(features[0][0])

        linear_output = self.linear(features).squeeze()
        #print(linear_output.size())

        out = torch.sigmoid(linear_output)
        #print(out)

        return out

    def get_num_class(self):
        return self.num_class

    def get_pretrain_model_name(self):
        return "MLP"
```

```python
model = mlp_classifier(use_lstm = False, mode = "CESL").to(device)
loss_fn = torch.nn.CrossEntropyLoss(ignore_index = -1).to(device)
optimizer = torch.optim.Adam(model.parameters(), lr = LEARNING_RATE)
```

```python
from tqdm.notebook import tqdm
from sentence_transformers import SentenceTransformer, util
from torch.utils.data import Dataset, DataLoader
from datasets import load_dataset, concatenate_datasets
import datasets

class commonsenseQA_dataset(Dataset):
    def __init__(self, datatype, tokenizer, local_file = None, sentence_model = None, choice_only = False):
        self.name = "commonsense_qa"
        self.datatype = datatype
        self.tokenizer = tokenizer
        self.num_choices = 5
        self.choice_only = choice_only
        self.sentence_model = sentence_model

        if local_file:
            print("Loading from local file", local_file)
            with open(local_file) as jsonfile:
                data = json.load(jsonfile)
                self.dataset = datasets.Dataset.from_dict(data)
        else:
            self.dataset = load_dataset(self.name, split = datatype)


            choice_feature = []
            embedding_feature = []
            similarity_feature = [] # cosine similarity between each choices (num_class * (num_class - 1))
            length_feature = []

            pbar = tqdm(total = self.dataset.num_rows, desc = "Generate New Feature")
            for d in range(self.dataset.num_rows):
                choice_feature.append([[1 if i == j else 0 for j in range(self.num_choices)] for i in range(self.num_choices)])
                embedding = [sentence_model.encode(text, convert_to_tensor=False) for text in self.dataset[d]["choices"]["text"]]
                cossim = [[] for _ in range(self.num_choices)]
                for i in range(self.num_choices):
                    for j in range(i, self.num_choices):
                        if i == j: continue
                        c = util.pytorch_cos_sim(embedding[i], embedding[j]).squeeze().tolist()
                        cossim[i].append(c)
                        cossim[j].append(c)

                embedding_feature.append(embedding)
                similarity_feature.append(cossim)
                length_feature.append([len(text) for text in self.dataset[d]["choices"]["text"]])

                pbar.update(1)
            pbar.close()

            # print(len(choice_feature), choice_feature[0])
            # print(len(embedding_feature), embedding_feature[0])
            # print(len(similarity_feature), similarity_feature[0])
            # print(len(length_feature), length_feature[0])

            new_features = {
                "choice_feature": choice_feature,
                "embedding_feature": embedding_feature,
                "similarity_feature": similarity_feature,
                "length_feature": length_feature,
            }

            new_ds = datasets.Dataset.from_dict(new_features)
            self.dataset = concatenate_datasets([self.dataset, new_ds], axis=1)

            # print(self.dataset.features)

        self.content = self.dataset.features.items()

    def __len__(self):
        return len(self.dataset)

    def __getitem__(self, index):
        return self.dataset[index]

    def num_classes(self):
        return 5

    def collate_fn(self, samples):
        num_choices = self.num_choices
        id = []
        question = []
        labels = []
        choices = []
        answer = []
        tokenized = []
        first_sentences = []
        second_sentences = []
        choice_feature = []
        embedding_feature = []
        similarity_feature = []
        length_feature = []

        for sample in samples:
            fsent = [sample["question"]] * num_choices
            ssent = [text for text in sample["choices"]["text"]]

            id.append(sample["id"])
            question.append(sample["question"])
            choices.append(sample["choices"]["text"])
            if "answerKey" in sample:
                answer.append(sample["answerKey"])
                labels.append(sample["choices"]["label"].index(sample["answerKey"]))

            # For tokenized
            first_sentences.append(fsent)
            second_sentences.append(ssent)

            # For more feature
            choice_feature.append(sample["choice_feature"])
            embedding_feature.append(sample["embedding_feature"])
            similarity_feature.append(sample["similarity_feature"])
            length_feature.append(sample["length_feature"])

        first_sentences = sum(first_sentences, [])
        second_sentences = sum(second_sentences, [])

        tokens = self.tokenizer(
            first_sentences,
            second_sentences,
            padding = True,
            max_length = MAX_LEN,
            truncation = True
        )

        if self.choice_only:
            tokens = self.tokenizer(second_sentences, padding = True, truncation = True)

        tokenized = {k: torch.tensor([v[i : i + num_choices] for i in range(0, len(v), num_choices)]) for k, v in tokens.items()}
        labels = torch.tensor(labels)
        choice_feature = torch.tensor(choice_feature)
        embedding_feature = torch.tensor(embedding_feature)
        similarity_feature = torch.tensor(similarity_feature)
        #similarity_feature = torch.mean(similarity_feature, axis = 2).reshape([-1, num_choices, 1])
        length_feature = torch.tensor(length_feature).reshape([-1, num_choices, 1])

        return{
            "id": id,
            "question": question,
            "labels": labels,
            "choices": choices,
            "answer": answer,
            "tokenized": tokenized,
            "choice_feature": choice_feature,
            "embedding_feature": embedding_feature,
            "similarity_feature": similarity_feature,
            "length_feature": length_feature
        }

    def get_name(self):
        ret = self.name + '_' + self.datatype
        if self.choice_only: ret += '_choice_only'
        return ret

    def get_dataset(self):
        return self.dataset
```

# Train

```python
from tqdm.notebook import tqdm

def train(dataloader, model, loss_fn, optimizer, device):
    train_loss, correct, total = 0, 0, 0

    model.train()
    pbar = tqdm(total = len(dataloader), desc = "Train")
    for batch, data in enumerate(dataloader):
        input_ids = data["tokenized"]["input_ids"].to(device)
        attention_mask = data["tokenized"]["attention_mask"].to(device)
        token_type_ids = data["tokenized"]["token_type_ids"].to(device)
        choice_feature = data["choice_feature"].to(device)
        embedding_feature = data["embedding_feature"].to(device)
        similarity_feature = data["similarity_feature"].to(device)
        length_feature = data["length_feature"].to(device)
        #print(data["choices"])

        labels = data['labels'].to(device)

        # bert
        # pred = model(input_ids, attention_mask, token_type_ids)

        # mlp
        pred = model(choice_feature, embedding_feature, similarity_feature, length_feature)

        # mix
        # pred = model(input_ids, attention_mask, token_type_ids, choice_feature, similarity_feature, length_feature)

        #print('pred', pred)

        loss = loss_fn(pred, labels)
        loss = loss / ACCUMULATION_STEPS
        train_loss += loss.item()

        loss.backward()
        if (batch + 1) % ACCUMULATION_STEPS == 0:
            optimizer.step()
            model.zero_grad()

        pred = pred.argmax(1)
        correct += torch.sum(pred == labels).item()
        total += labels.numel() - torch.sum(labels == -1).item()

        #print(pred, labels, torch.sum(pred == labels).item(), labels.numel() - torch.sum(labels == -1).item())

        pbar.update(1)
        if batch > 0 and batch % 1000 == 0:
            print("Step:", batch, "Accuracy:", 100 * correct / total, "%")
            #print(correct, total)
            #break


    pbar.close()

    train_loss /= len(dataloader)
    print(f"Train: Accuracy: {(100 * correct / total):>0.1f}%, Avg loss: {train_loss:>8f} \n")

    return correct, total
```

# Evaluate

```python
from tqdm.notebook import tqdm

def eval(dataloader, model, loss_fn, device):
    eval_loss, correct, total = 0, 0, 0

    model.eval()
    pbar = tqdm(total = len(dataloader), desc = "Eval")
    for batch, data in enumerate(dataloader):
        model.zero_grad()
        input_ids = data["tokenized"]["input_ids"].to(device)
        attention_mask = data["tokenized"]["attention_mask"].to(device)
        token_type_ids = data["tokenized"]["token_type_ids"].to(device)
        choice_feature = data["choice_feature"].to(device)
        embedding_feature = data["embedding_feature"].to(device)
        similarity_feature = data["similarity_feature"].to(device)
        length_feature = data["length_feature"].to(device)
        #print(data["choices"])

        labels = data['labels'].to(device)

        # bert
        # pred = model(input_ids, attention_mask, token_type_ids)

        # mlp
        pred = model(choice_feature, embedding_feature, similarity_feature, length_feature)

        # mix
        # pred = model(input_ids, attention_mask, token_type_ids, choice_feature, similarity_feature, length_feature)

        #print('pred', pred)

        loss = loss_fn(pred, labels)
        eval_loss += loss.item()

        pred = pred.argmax(1)

        correct += torch.sum(pred == labels).item()
        total += labels.numel() - torch.sum(labels == -1).item()

        pbar.update(1)
        if batch > 0 and batch % 1000 == 0:
            print("Step:", batch,"Accuracy:", 100 * correct / total, "%")
            #print(correct, total)
            #break

    pbar.close()

    eval_loss /= len(dataloader)
    print(f"Eval: Accuracy: {(100 * correct / total):>0.1f}%, Avg loss: {eval_loss:>8f} \n")

    return correct, total
```
